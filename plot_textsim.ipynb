# Load dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("./results/similarity_results.csv")
```

```python
# Mean plot scores
mean_scores = df.groupby("Model")[["Train pearson", "Val pearson", "Test pearson"]].mean()
mean_scores = mean_scores.sort_values(ascending=False, by="Test pearson")
mean_scores.plot(kind="bar", figsize=(12, 6))
plt.title("Mean Scores for Each Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=55)
plt.legend(title="Scores")
plt.show()

# Mean validation scores
mean_scores = df.groupby("Model")[ "Val pearson"].mean()
mean_scores = mean_scores.sort_values(ascending=False)
mean_scores.plot(kind="bar", figsize=(12, 6))
plt.title("Validation Scores for Each Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=30)
plt.legend(title="Scores")
plt.show()

# Max pearson scores
max_scores = df.groupby("Model")[["Train pearson", "Val pearson", "Test pearson"]].max()
max_scores = max_scores.sort_values(ascending=False, by="Test pearson")
max_scores.plot(kind="bar", figsize=(12, 6))
plt.title("Best Scores for Each Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=55)
plt.legend(title="Scores")
plt.show()

# Boxplot of pearson scores
df_melted = df.melt(id_vars=["Model"], value_vars=["Train pearson", "Val pearson", "Test pearson"], var_name="Score Type", value_name="Score")
plt.figure(figsize=(12, 6))
ax = plt.gca()
df_melted.boxplot(by="Model", column="Score", grid=False, ax=ax)
plt.title("Score Distributions for Each Model")
plt.suptitle('')
plt.xlabel("Model")
plt.ylabel("Score")
plt.xticks(rotation=30)
plt.show()

# Scores for each execution model
mean_scores = df.groupby("Exec model")[["Train pearson", "Val pearson", "Test pearson"]].mean()
mean_scores.plot(kind="bar", figsize=(12, 6))
plt.title("Mean Scores for Each Tensorflow Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=30)
plt.legend(title="Scores")
plt.show()
```

```python
max_scores = df.groupby("Model")[["Train spearman", "Val spearman", "Test spearman"]].max()
max_scores = max_scores.sort_values(ascending=False, by="Test spearman")
max_scores.plot(kind="bar", figsize=(12, 6))
plt.title("Best Spearman Scores for Each Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=55)
plt.legend(title="Scores")
plt.show()
```

### Word2Vec

```python
word2vec_df = df[df["Model"].str.startswith("Word2vec")]

# Order
mean_test_pearson = word2vec_df.groupby("Model")["Test pearson"].mean()
max_scores_word2vec = word2vec_df.groupby("Model")["Test pearson"].max()
sorted_models = mean_test_pearson.sort_values(ascending=False).index
sorted_models_max = max_scores_word2vec.sort_values(ascending=False).index

word2vec_df["Model"] = pd.Categorical(word2vec_df["Model"], categories=sorted_models, ordered=True)
word2vec_df = word2vec_df.sort_values("Model")

mean_scores_word2vec = word2vec_df.groupby("Model")[["Train pearson", "Val pearson", "Test pearson"]].mean()
mean_scores_word2vec.plot(kind="bar", figsize=(12, 6))
plt.title("Mean Scores for Each Word2vec Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.legend(title="Scores")
plt.show()

word2vec_df["Model"] = pd.Categorical(word2vec_df["Model"], categories=sorted_models_max, ordered=True)
word2vec_df = word2vec_df.sort_values("Model")
max_scores_word2vec = word2vec_df.groupby("Model")[["Train pearson", "Val pearson", "Test pearson"]].max()
max_scores_word2vec.plot(kind="bar", figsize=(12, 6))
plt.title("Best Scores for Each Word2vec Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.legend(title="Scores")
plt.show()

word2vec_df["Model"] = pd.Categorical(word2vec_df["Model"], categories=sorted_models_max, ordered=True)
word2vec_df = word2vec_df.sort_values("Model")
max_scores_word2vec = word2vec_df.groupby("Model")[["Train spearman", "Val spearman", "Test spearman"]].max()
max_scores_word2vec.plot(kind="bar", figsize=(12, 6))
plt.title("Best Spearman Scores for Each Word2vec Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.legend(title="Scores")
plt.show()

word2vec_melted = word2vec_df.melt(id_vars=["Model"], value_vars=["Train pearson", "Val pearson", "Test pearson"], var_name="Score Type", value_name="Score")
plt.figure(figsize=(12, 6))
ax = plt.gca()
word2vec_melted.boxplot(by="Model", column="Score", grid=False, ax=ax)
plt.title("Score Distributions for Each Word2vec Model")
plt.suptitle('')
plt.xlabel("Model")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.show()

mean_scores_word2vec = word2vec_df.groupby("Exec model")[["Train pearson", "Val pearson", "Test pearson"]].mean()
mean_scores_word2vec.plot(kind="bar", figsize=(12, 6))
plt.title("Mean Scores for Each Tensorflow Model")
plt.ylabel("Score")
plt.xlabel("Exec model")
plt.xticks(rotation=45)
plt.legend(title="Scores")
plt.show()
```

### Trainable embeddings

```python
word2vec_df = df[df["Mode"] == "embeddings"]

# Calculate mean Test Pearson scores for each model
mean_test_pearson = word2vec_df.groupby("Model")["Test pearson"].mean()
max_scores_word2vec = word2vec_df.groupby("Model")["Test pearson"].max()
# Sort models by mean Test Pearson scores in descending order
sorted_models = mean_test_pearson.sort_values(ascending=False).index
sorted_models_max = max_scores_word2vec.sort_values(ascending=False).index

# Reorder the DataFrame based on the sorted models
word2vec_df["Model"] = pd.Categorical(word2vec_df["Model"], categories=sorted_models, ordered=True)
word2vec_df = word2vec_df.sort_values("Model")

# Plotting
# 1. Mean results of each Word2vec model
mean_scores_word2vec = word2vec_df.groupby("Model")[["Train pearson", "Val pearson", "Test pearson"]].mean()
mean_scores_word2vec.plot(kind="bar", figsize=(12, 6))
plt.title("Mean Scores for Each Embedding Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.legend(title="Scores")
plt.show()



word2vec_df["Model"] = pd.Categorical(word2vec_df["Model"], categories=sorted_models_max, ordered=True)
word2vec_df = word2vec_df.sort_values("Model")


# 2. Best results of each Word2vec model
max_scores_word2vec = word2vec_df.groupby("Model")[["Train pearson", "Val pearson", "Test pearson"]].max()
max_scores_word2vec.plot(kind="bar", figsize=(12, 6))
plt.title("Best Scores for Each Embedding Model")
plt.ylabel("Score")
plt.xlabel("Model")
plt.xticks(rotation=45)
plt.legend(title="Scores")
plt.show()

# 3. Box plot for score distributions of Word2vec models
word2vec_melted = word2vec_df.melt(id_vars=["Model"], value_vars=["Train pearson", "Val pearson", "Test pearson"], var_name="Score Type", value_name="Score")
plt.figure(figsize=(12, 6))
ax = plt.gca()
word2vec_melted.boxplot(by="Model", column="Score", grid=False, ax=ax)
plt.title("Score Distributions for Each Embedding Model")
plt.suptitle('')
plt.xlabel("Model")
plt.ylabel("Score")
plt.xticks(rotation=45)
plt.show()



mean_scores_word2vec = word2vec_df.groupby("Exec model")[["Train pearson", "Val pearson", "Test pearson"]].mean()
mean_scores_word2vec.plot(kind="bar", figsize=(12, 6))
plt.title("Mean Scores for Each Tensorflow Model")
plt.ylabel("Score")
plt.xlabel("Exec model")
plt.xticks(rotation=45)
plt.legend(title="Scores")
plt.show()
```
